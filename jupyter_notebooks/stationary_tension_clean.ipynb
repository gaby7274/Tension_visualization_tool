## first section for stationary harmony, 
usan algo que se llama T_sub, esto significa el common subharmonic, o realmente T_sub es el periodo aproximado de la onda compuesta de varias notas. Es una combinación lineal de las diferentes frecuencias, pero para calcularlo, es un papelón. Ya que hay una fórmula particulat

La fórmula, T_sub = avg(k_i*t_i)
esta k_i es un integer multiplier al periodo de cada onda que se necesita para poder decir llegué. Proof read y leer fisica para esto, sin embargo, un ejemplo. 

para el periodo1 = 7.6ms y periodo2= 6.1ms, son las ondas C3 y E3. 
Estas ondas cuando están juntas, la primera da oscilaciones son estas.  

```python


def calculate_period_dataframe(note_names:list,note_df ,max=13):
    df = pd.DataFrame()
    for note_name in note_names:
        df[note_name]= 1/note_df[note_name][0]

    ## calculating period
    for i in range(1,max):
        list_of_periods = []
        for note_name in note_names:
            list_of_periods.append((1/(note_df[note_name][0]))*i)

        df.loc[i] = list_of_periods
    # note_period_c3 = (1/note_c3[0]) * (i)
    # note_period_e3 = (1/note_e3[0]) * (i)
    # note_period_g3 = (1/note_g3[0]) * (i)
    # note_period_c4 = (1/note_c4[0]) * (i)
    # dataframe_demos.loc[i] = [note_period_c3, note_period_e3, note_period_g3, note_period_c4]


    
    return df

note_names = ['C3', 'D#3', 'G3', 'A#3']
period_df = calculate_period_dataframe(note_names, fundamental_df)  
period_df

```

```python
import numpy as np


def create_note_difference_matrix(period_df):
    notes = period_df.columns
    ranges = []
    values = [] 
    shape = []
    for note in notes:
        # ranges.append(note)
        values.append(period_df[note].values)
        shape.append(len(period_df[note].values))
    matrix = np.zeros(shape)
    
    
    ## works for for arrays only
    array1 = values[0]
    array2 = values[1]
    array3 = values[2]
    array4 = values[3]
    for i in range(len(array1)):
        for j in range(len(array2)):
            for k in range(len(array3)):
                for l in range(len(array4)):
                    ## find maximum difference between the four notes at each i,j,k,l
                    min_num = min(array1[i], array2[j], array3[k], array4[l])
                    max_num = max(array1[i], array2[j], array3[k], array4[l])

                    matrix[i,j,k,l] = max_num - min_num




    
    return matrix
           
```

```python


def increment_index(amount_of_notes, range_considered):
    """ 
    amount_of_notes: array of indexes, whose length is equal of notes played
    range_considered: int of maximum, range to consider index
    """ 
    index_to_increment = len(amount_of_notes) - 1
    incrementing = 1
    loops = 0
    while incrementing and index_to_increment>=0:
        amount_of_notes[index_to_increment]+=1
        loops += 1
            # print('loops:', loops)
        if amount_of_notes[index_to_increment] >= range_considered:
            amount_of_notes[index_to_increment] = 0
            index_to_increment -= 1
        else:
            incrementing = 0

    return amount_of_notes, index_to_increment, loops

amount_of_notes = [0,0,0,0]
range_considered = 15
index_to_increment = len(amount_of_notes) - 1

while index_to_increment >= 0:
    amount_of_notes, index_to_increment, loops = increment_index(amount_of_notes, range_considered)
    print(amount_of_notes, index_to_increment, loops)




```

### Ahora, intentaremos hacer algo loco, encontrar las cotas menores y mayores
Algoritmo es T(n)=2n

```python
def finding_min_difference_with_lower_upper_bound(period_df):
    '''
    This function will assume that the first and last note are the lower and upper bounds
    Since bass is the first note, and it oscillates less, than the last note

    Algorithm goes as follows:
    loop until last note max index is reached

    then, for each note, we will take the difference between the last note and the bass note
    and divide by the mean of the two notes. We will store, the keys, the division and 
    then sort by division results. 

    each time the difference of last-bass is positive we will increment the bass,
    then when the difference is negative, we will increment the last note
    '''
    notes = period_df.columns
    range_to_consider = len(period_df[notes[0]].values)

    ## bass
    bass =  period_df[notes[0]].values

    last_note = period_df[notes[-1]].values

    ## algorithm
    i, j = 0,0

    ## array to store all dicts info
    ratio_dicts = []

    while i < range_to_consider and j < range_to_consider:
        bass_note = bass[i]
        last_note_value = last_note[j]

        difference = last_note_value - bass_note
        mean = (last_note_value + bass_note) / 2
        ratio = abs(difference/mean)
        ratio_dicts.append({
            'bass': i,
            'last_note': j,
            'difference': difference,
            'mean': mean,
            'ratio': ratio,
            'bass_value':bass_note,
            'last_note_value':last_note_value
        })

        if difference > 0:
            i += 1
        else:
            j += 1
        
    return pd.DataFrame(sorted(ratio_dicts, key=lambda x: x['ratio'] ))
```

```python

def binary_search_with_bounds(period_df,lower_bound,upper_bound,mean_bound):
    

    """
    this binary search goes as follows:
    1. We first take a pivot, which is between lower and upper bound,
    2. we then substract period against mean. 
    3. we compare if that is the lowest, if it is, take pivot as candidate
    4. if positive, we go to right leaf, if negative we go to left
    5. at the end return candidate for each 
    """

    notes = period_df.columns
    ## only take medium ones
    notes = notes[ 1 : -1]

    chosen_winners = []
    all_found = []
    for note in notes:

        ### here we begin binary loop

        low = lower_bound
        upper = upper_bound
        
        candidate_index = 0
        candidate_difference = 1000
        candidate_value = 10000
        # updated= True
        note_values = period_df[note].values
        # print(note_values)


        complete_note_differences = []

        while low<=upper:
            # updated = False

            middle = (low + upper) // 2

            difference = mean_bound - note_values[middle]


            if(candidate_difference>abs(difference)):
                candidate_difference =  abs(difference)
                candidate_index = middle
                candidate_value = note_values[middle]
            complete_note_differences.append({
                'index':middle,
                'value': note_values[middle],
                'abs_difference': abs(difference),
                'difference':difference
            })
            ## update middle
            if difference >0:
                low=middle+1
            else:
                upper = middle-1

        #### now we store the necessary information, the chosen one and everything
        all_found.append(
            {'note': note,
             'values':complete_note_differences}
        )
        chosen_winners.append(
            {'note': note,
             'candidate_index': candidate_index,
             'candidate_difference':candidate_difference,
             'candidate_value': candidate_value}
        )
    return all_found,chosen_winners

            
            
        

def choosing_bounds_binary_search(period_df, bounds_df):
    
    # notes in period_df to get bass and last
    notes = period_df.columns
    # bass =  bounds_df[notes[0]].values
    # last_note =  bounds_df[notes[-1]].values
    each_pair_values = []
    for i,row in bound_df.iterrows():
        bass_value = int(row['bass'])
        last_note_value = int(row['last_note'])
        mean_bound = row['mean']
        all_found, chosen_winners = binary_search_with_bounds(period_df,bass_value,last_note_value, mean_bound)
        

        ### interestingly enough, could i then here, normalize, and standard deviation?
        values=[row['bass_value']]
         ## indeces
        indices_to_observe=[row['bass']]
        for d in chosen_winners:
            values.append(d['candidate_value'])
            indices_to_observe.append(d['candidate_index'])
        indices_to_observe.append(row['last_note'])
        values.append(row['last_note_value'])


       


        ## normalizing, and calculating standard deviation and mean
        

        minimum = min(values)
        maximum = max(values)
        mean_vals = sum(values)/len(values)

        ##normalizing
        norm_vals = []
        for v in values:
            normed = (v-minimum)/(maximum-minimum)
            norm_vals.append(normed)
        std_= statistics.stdev(norm_vals)

        observable_metrics = {
            't_values':values,
            'mean_t_sub': mean_vals,
            'std_normed':std_,
            'indices':indices_to_observe,
            'difference':maximum-minimum,
            'difference_ratio': (maximum-minimum)/mean_vals,
        }

        




        each_pair_values.append({
            "all_found":all_found,
            'chosen_winners':chosen_winners,
            'observable_metrics':observable_metrics
            
        }
        )
        
    return each_pair_values

```

```python

# Sample frequency array (in Hertz)
frequencies =fundamental_df[note_names].to_numpy().flatten()  # C4, E4, G4 notes

# Function to generate sine wave data
def generate_sine_waves(frequencies, duration=1.0, sample_rate=1000):
    t = np.linspace(0, duration, int(sample_rate * duration))
    data = []
    
    for freq in frequencies:
        # Calculate sine wave values: sin(2π * frequency * time)
        y_values = np.sin(2 * np.pi * freq * t)
        
        # Create DataFrame entries for this frequency
        for i in range(len(t)):
            data.append({
                'time': t[i],
                'amplitude': y_values[i],
                'frequency': f"{freq} Hz"
            })
            
    return pd.DataFrame(data)

# Generate the data
sine_df = generate_sine_waves(frequencies,duration=((1/frequencies[-1])*18), sample_rate=10000)

# Create Altair visualization
chart = alt.Chart(sine_df).mark_line(
    opacity=0.5
).encode(
    x=alt.X('time:Q', title='Time (seconds)'),
    y=alt.Y('amplitude:Q', title='Amplitude'),
    color=alt.Color('frequency:N', title='Frequency')
).properties(
    width=600,
    height=300,
    title='Sine Waves at Different Frequencies'
)

rect_chart = alt.Chart(observable_df).mark_rect(
    opacity=0.5
).encode(
    x='t_value_min',
    x2='t_value_max',
    # color='order'
)
chart+rect_chart
```

### Comenzaré idea interesante. Vamos a ver, vamos a graficar los deltas de progresiones, igual, escoger los delta_t donde la diferencia de Tsubs es más pequeñas, y ver como se comportan. 

```python
def calculate_nearest_tsubs(previous_df,next_df):
    index_previous=0
    difference_rows={
        'shortest_t_sub':[],
        'shortest_t_sub_index':[],
        'delta_delta_t':[]

    }
    for i,r in next_df.iterrows():
        smallest_tsub=9999
        tsub_index = 0
        delta_delta_t = 0
        shortest_t_sub=0
        while index_previous<len(previous_df):

            ## substract totals...
            t_sub_dif = r['mean_t_sub'] - previous_df.iloc[index_previous]['mean_t_sub']

            ## if possible smallest encountered save possible candidate
            if (smallest_tsub>abs(t_sub_dif)):
                smallest_tsub= t_sub_dif
                shortest_t_sub =  previous_df.iloc[index_previous]['mean_t_sub']
                index_previous +=1
            ## if smallest_tsub is smaller than the difference then the smallest was the previous
            else:
                
                break
        index_previous-=1
        difference_rows['shortest_t_sub'].append( shortest_t_sub)
        difference_rows['shortest_t_sub_index'].append(index_previous)
        difference_rows['delta_delta_t'].append(previous_df.iloc[index_previous]['difference']-r['difference'])
    return difference_rows
```

```python



def calculate_total_df(chord_transitions,chord_progression_name):
    total_df = pd.DataFrame()
    previous_df=None


    for index,chord in enumerate(chord_transitions):
        print(index,chord)
        period_df = calculate_period_dataframe(chord,fundamental_df,max=20)
        bound_df = finding_min_difference_with_lower_upper_bound(period_df)
        each_pair = choosing_bounds_binary_search(period_df,bound_df)
        ## we will see if our method seems ok????
        observable_metrics=[]
        for item in each_pair:
            observable_metrics.append(item['observable_metrics'])
        sorted(observable_metrics, key=lambda x: x['difference'])
        top_3_observable_metrics = sorted(observable_metrics, key=lambda x: x['difference'])[:3]
    
        observable_df = pd.DataFrame(top_3_observable_metrics)
        observable_df.sort_values(by='mean_t_sub',inplace=True)
        observable_df['chord_num'] = chord_progression_name[index]
        observable_df['chord'] = [chord]*3
        observable_df['chord_index'] = [index]*3
        ## adding 1,2,3, order values to sorted by mean
        observable_df['order'] = [1,2,3]
        observable_df['ratio_delta_t_tsub']= observable_df['difference']/observable_df['mean_t_sub']
        # observable_df.iloc[0]

        if previous_df is None:
            total_df = observable_df
            # total_df['ratio_delta_t_tsub']= total_df['difference']/total_df['mean_t_sub']
            total_df['delta_delta_t'] = 0
            total_df['shortest_t_sub']= 0
            total_df['shortest_t_sub_index'] = 0
            total_df['delta_t_tsub_ratio']=0

            previous_df = total_df
        else:
            ## this contains smallest tsubs from the succeeding chords
            ## the delta_delta_t
            ## the tsub-index...
            difference_dictionary = calculate_nearest_tsubs(previous_df,observable_df)
            
            observable_df['delta_delta_t'] = difference_dictionary['delta_delta_t']
            observable_df['shortest_t_sub'] = difference_dictionary['shortest_t_sub']
            observable_df['shortest_t_sub_index'] = difference_dictionary['shortest_t_sub_index']
            observable_df['delta_t_tsub_ratio'] = observable_df['delta_delta_t']/observable_df['mean_t_sub']

            total_df = pd.concat([total_df,observable_df],ignore_index=True)
            previous_df = observable_df

            ##line graphs of tsubs
    total_df.fillna(0, inplace=True)


    return total_df







    
```
