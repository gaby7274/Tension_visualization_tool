```python


#hz is the number
def equation_for_midi_number(hz):
    return 12*log2(hz/440) +69
    


# N is themidi number

def equation_for_hertz(n):
    return 440* 2**((n-69)/12)

def equation_for_harmonics(number_of_harmonics, midi_number):
    all_harmonics = []
    hertz = equation_for_hertz(midi_number)
    for i in range(number_of_harmonics):
        
        all_harmonics.append((i+1)*hertz)
    
    return all_harmonics
        

note_df = pd.DataFrame()


piano_range = range(21, 109)
piano_notes = ['A', 'A#', 'B', 'C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#']
piano_octave = [str(i) for i in range(0, 9)]

##populating dataframe columns
piano_note_index = 0
piano_octave_index = 0
for midi_number in piano_range:
    if piano_note_index %12 == 3:
        piano_octave_index += 1

    note_name = piano_notes[piano_note_index % 12] + piano_octave[piano_octave_index]
    note_df[note_name]= equation_for_harmonics(10,midi_number)
    
    piano_note_index += 1


note_df

    
```

## Intento final de dissonance score

<ul>
<li>2^0 === consonant</li>
<li>2^1 === Superdissonant</li>
<li>2^2=== Moredissonant</li>
<li>2^3 === somewhatDissonant</li>
<li>2^4 === somewhatDis</li>
<li>2^5 === disonant. more dissonant</li>
<li>2^6 === consonant</li>
<li>2^7 === moreDissonant</li>
<li>2^8 === somewhatDissonant</li>
<li>2^9 === somewhatDissonant</li>
<li>2^10 === moreDissonant</li>
<li>2^11 === superDissonant</li>
</ul>

```python

def calculate_tension_score(list_of_values):
    consonant_ratio = -1
    somewhat_dis = 0.33
    moreDis = 0.66
    Dissonant =2

    # so we say that the first index is consonant, and the [6]
    #consonancy 
    tension_score = consonant_ratio *list_of_values[0] + consonant_ratio * list_of_values[6]
    #somewhat_dis
    tension_score+= (somewhat_dis * (list_of_values[3] +list_of_values[4] + list_of_values[8] + list_of_values[9]))
    # moreDis
    tension_score+= (moreDis *(list_of_values[2]+list_of_values[7]+list_of_values[10] +list_of_values[5]))
    tension_score += (Dissonant *(list_of_values[1]+ list_of_values[11]))
    tension_score /= sum(list_of_values)
    return tension_score

```
