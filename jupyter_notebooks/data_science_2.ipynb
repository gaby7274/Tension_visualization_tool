```python
import pandas as pd
from math import log2
import numpy as np




#hz is the number
def equation_for_midi_number(hz):
    return 12*log2(hz/440) +69
    


# N is themidi number

def equation_for_hertz(n):
    return 440* 2**((n-69)/12)

def equation_for_harmonics(number_of_harmonics, midi_number):
    all_harmonics = []
    hertz = equation_for_hertz(midi_number)
    for i in range(number_of_harmonics):
        
        all_harmonics.append((i+1)*hertz)
    
    return all_harmonics
        

note_df = pd.DataFrame()


piano_range = range(21, 109)
piano_notes = ['A', 'A#', 'B', 'C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#']
piano_octave = [str(i) for i in range(0, 9)]

##populating dataframe columns
piano_note_index = 0
piano_octave_index = 0
for midi_number in piano_range:
    if piano_note_index %12 == 3:
        piano_octave_index += 1

    note_name = piano_notes[piano_note_index % 12] + piano_octave[piano_octave_index]
    note_df[note_name]= equation_for_harmonics(10,midi_number)
    
    piano_note_index += 1


note_df

    
```

## Esta secciÃ³n lo que hace:

1. por cada uno de los armÃ³nicos, se restan contra los otros (Ai-Bj, donde i es armonico de A y j armonico B)
2. Donde estÃ¡n esas restas, y se encuentran menor que 170, le sacamos el promedio de las dos ondas
3. Lo que estaremos evaluado entonces es, cuan tensa se vuelve la onda promedio (F_promedio) entre las dos al modularla con (Ai-Bj)hz
4. Una de las generalizaciones que hacemos es, cuando delta(F_k) = (Ai-Bj)<=170 y  delta(F_k) =(2^(1/12)-1) F_promedio, entonces la onda promedio <br>
    a esa modulaciÃ³n es disonante.
5. (usa este link https://academo.org/demos/amplitude-modulation/, usa f1 a 440 con 1Hz, 2hz, y luego 26 hz donde 26 =~~ (2^(1/12))*440, y coincide con la distancia de un 440 y un semitono. Con 2^(2/12) es la de un tono
6. Luego, lo que calculamos en esto es todas las ondas en las cuales lower_bound<delta(Fk)/F_promedio< high_bound.
7. Los valores de los bounds son low = 2^(g/12), high = 2^(g+1/12), for g in [0.0,0.125,0.25,0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5,8.5,9.5,10.5,11.5,12.5]
8. note que entre 0, y 0.125 son los ranges sumamente consonantes y pertenecen a la 8va. 0.125 a 0.25 es un pocomÃ¡s disonante, pero los que se
9. consideran son aquellos mÃ¡s disonantes, (segun elpaper, central region of disonance), y tambiÃ©n los puntos que caen en rojo

```python



no_weights = False


def tension_df(tension_list, values_of_r_ranges):

    ranges_dataframe = pd.DataFrame()
    ranges_dataframe['interval']=['U', 'm2', 'M2', 'm3', 'M3', 'P4', 'TT', 'P5', 'm6', 'M6', 'm7', 'M7', '8v']
    df_ten = pd.DataFrame(tension_list)
    for i in range(len(values_of_r_ranges)-1):
        ranges_dataframe[f'2**({values_of_r_ranges[i]}/12)<ratio<2**{values_of_r_ranges[i+1]}/12'] = (df_ten)[i]
    
    return ranges_dataframe.T




def calculate_tension_v2(note1, note2, note_df, values_of_r, no_weights=False, no_limit=False):

    restas = pd.DataFrame()
    puntos_medios = pd.DataFrame()
    for i in range(10):
        restas[str(i)+'th']=  abs(note_df[note1][i] -note_df[note2]) 
        puntos_medios[str(i)+'th'] = abs((note_df[note1][i] + note_df[note2])/2)

    #  filtramos por threshold de batimientos??

    if no_limit is False:
        restas =  restas[restas <=170]
    
    # print(restas)
    # print(puntos_medios)

    # return restas
    # print(restas)
    ratio_df = restas/puntos_medios

    # lets see sum for each upper bound
    dif_ranges = values_of_r
    ranges_list = []
    offsets= (0.1,0.1)

    ## Each element inside ranges_list is all the harmonics that are
    # bounded by lower and upper
    for i in range(len(dif_ranges)-1):
        lower_bound = ratio_df >= 2**((dif_ranges[i])/12)-1
        upper_bound = ratio_df < 2**((dif_ranges[i+1])/12)-1

        # first .sum() yields amount of modulations in each harmonic
        # that fall in those boundaries. 

        ##  Let note1 be sum(Fi) and note2 sum(Fj)
        # then we calculates F[i] - allF[j] for each harmonic_i,
        # and  then .sum() gives amount of modulations that fall in those boundaries
        # for each harmonic_i
        weights = np.array([len((lower_bound & upper_bound).sum())] *len((lower_bound & upper_bound).sum())) / ((np.array(range(1, len((lower_bound & upper_bound).sum())+1)))*10)
       
        
        if(no_weights):
            weights = np.array([1] *len((lower_bound & upper_bound).sum()))
        


        
        weighted = (lower_bound & upper_bound).sum() *weights
        
        

        #for purposes, the second.sum() calculates total modulation in each boundary within that interval

        
        ranges_list.append(weighted.sum())

    return ranges_list



# for offsets in offset_list:

#     tension_list =[calculate_tension_v2('G3',note, note_df, offsets) for note in notes_to_analyze]
#     print(tension_df(tension_list, offsets))



notes_to_analyze = ['G3','G#3','A3', 'A#3', 'B3', 'C4', 'C#4', 'D4', 'D#4', 'E4', 'F4', 'F#4','G4']

print('\n\nwithout weights\n\n')
no_weights = False

values_of_r_ranges = [0.0,0.125,0.25,0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5,8.5,9.5,10.5,11.5,12.5]

# for offsets in offset_list:

#     tension_list =[calculate_tension_v2('G3',note, note_df, offsets, values_of_r_ranges) for note in notes_to_analyze]
#     print(tension_df(tension_list, values_of_r_ranges))

# # tension_list_unfiltered = [calculate_tension_v1_test('G3',note, note_df) for note in notes_to_analyze]

no_weights = True
# for offsets in offset_list:

tension_list =[calculate_tension_v2('G3',note, note_df, values_of_r_ranges) for note in notes_to_analyze]
print(tension_df(tension_list, values_of_r_ranges))

```

```python
offsets= (0.6,0.4)
no_weights = False
notes_to_analyze = ['G3','G#3','A3', 'A#3', 'B3', 'C4', 'C#4', 'D4', 'D#4', 'E4', 'F4', 'F#4','G4' ]
intervals = ['U', 'm2', 'M2', 'm3', 'M3', 'P4', 'TT', 'P5', 'm6', 'M6', 'm7', 'M7','P8' ]
values_of_r_ranges = [0.0,0.125,0.25,0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5,8.5,9.5,10.5,11.5,12.5]
tension_list =[calculate_tension_v2('G3',note, note_df, values_of_r_ranges) for note in notes_to_analyze]
print(tension_df(tension_list, values_of_r_ranges))

print(tension_list)

```

<ul>
<li>2^0 <= f < 2^0.125 === consonant</li>
    <li>2^0.125 <= f < 2^0.25 === chinita-more  dis</li>
        <li>2^0.25 <= f < 2^0.5 ===chinita-more dis </li>
            <li>2^0.5 <= f < 2^1.5 === dissonant</li>

 <li>2^1.5 <= f < 2^2.5 === dissonant</li>
      <li>2^2.5<= < f < 2^3.5 ===more_dissonant</li>
           <li>2^3.5 < f < 2^4.5 === ssomewhat_dis</li>
                <li>2^4.5 < f < 2^5.5 === dissonant</li>

 <li>2^5.5 <= f < 2^6.5 === somewhat</li>
 <li>2^6.5 <= f < 2^7.5 === somwhat</li>
 <li>2^7.5 <= f < 2^8.5 === more_dis</li>


```python

def calculate_tension_score(list_of_values):
    consonant_ratio = 0
    somewhat_dis = 1
    moreDis = 2
    Dissonant =3

    # so we say that the first index is consonant, and the [6]
    #consonancy 
    tension_score = consonant_ratio *(list_of_values[0] )
    #somewhat_dis
    tension_score+= (somewhat_dis * (list_of_values[6]+ list_of_values[8])+  list_of_values[9]) #+ list_of_values[6] + list_of_values[8]))
    # moreDis
    tension_score+= (moreDis *(list_of_values[1] +list_of_values[2]+ list_of_values[5]+list_of_values[10]))#+list_of_values[10] +list_of_values[5]))
    tension_score += (Dissonant *(list_of_values[3]+ list_of_values[4]+list_of_values[7]))
    tension_score /= sum(list_of_values)
    return tension_score

```

```python
interval_dict = {}
for note, interval, lista in zip(notes_to_analyze, intervals, tension_list):
    interval_dict[interval] = calculate_tension_score(lista)
sorted_dict =  sorted(interval_dict.items(), key=lambda x: x[1])
sorted_dict
```
