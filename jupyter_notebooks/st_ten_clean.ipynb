### Ahora, intentaremos hacer algo loco, encontrar las cotas menores y mayores
Algoritmo es T(n)=2n

```python
def finding_min_difference_with_lower_upper_bound(period_df):
    '''
    This function will assume that the first and last note are the lower and upper bounds
    Since bass is the first note, and it oscillates less, than the last note

    Algorithm goes as follows:
    loop until last note max index is reached

    then, for each note, we will take the difference between the last note and the bass note
    and divide by the mean of the two notes. We will store, the keys, the division and 
    then sort by division results. 

    each time the difference of last-bass is positive we will increment the bass,
    then when the difference is negative, we will increment the last note
    '''
    notes = period_df.columns
    range_to_consider = len(period_df[notes[0]].values)

    ## bass
    bass =  period_df[notes[0]].values

    last_note = period_df[notes[-1]].values

    ## algorithm
    i, j = 0,0

    ## array to store all dicts info
    ratio_dicts = []

    while i < range_to_consider and j < range_to_consider:
        bass_note = bass[i]
        last_note_value = last_note[j]

        difference = last_note_value - bass_note
        mean = (last_note_value + bass_note) / 2
        ratio = abs(difference/mean)
        ratio_dicts.append({
            'bass': i,
            'last_note': j,
            'difference': difference,
            'mean': mean,
            'ratio': ratio,
            'bass_value':bass_note,
            'last_note_value':last_note_value
        })

        if difference > 0:
            i += 1
        else:
            j += 1
        
    return pd.DataFrame(sorted(ratio_dicts, key=lambda x: x['ratio'] ))
```

```python

def binary_search_with_bounds(period_df,lower_bound,upper_bound,mean_bound):
    

    """
    this binary search goes as follows:
    1. We first take a pivot, which is between lower and upper bound,
    2. we then substract period against mean. 
    3. we compare if that is the lowest, if it is, take pivot as candidate
    4. if positive, we go to right leaf, if negative we go to left
    5. at the end return candidate for each 
    """

    notes = period_df.columns
    ## only take medium ones
    notes = notes[ 1 : -1]

    chosen_winners = []
    all_found = []
    for note in notes:

        ### here we begin binary loop

        low = lower_bound
        upper = upper_bound
        
        candidate_index = 0
        candidate_difference = 1000
        candidate_value = 10000
        # updated= True
        note_values = period_df[note].values
        # print(note_values)


        complete_note_differences = []

        while low<=upper:
            # updated = False

            middle = (low + upper) // 2

            difference = mean_bound - note_values[middle]


            if(candidate_difference>abs(difference)):
                candidate_difference =  abs(difference)
                candidate_index = middle
                candidate_value = note_values[middle]
            complete_note_differences.append({
                'index':middle,
                'value': note_values[middle],
                'abs_difference': abs(difference),
                'difference':difference
            })
            ## update middle
            if difference >0:
                low=middle+1
            else:
                upper = middle-1

        #### now we store the necessary information, the chosen one and everything
        all_found.append(
            {'note': note,
             'values':complete_note_differences}
        )
        chosen_winners.append(
            {'note': note,
             'candidate_index': candidate_index,
             'candidate_difference':candidate_difference,
             'candidate_value': candidate_value}
        )
    return all_found,chosen_winners

            
            
        

def choosing_bounds_binary_search(period_df, bounds_df):
    
    # notes in period_df to get bass and last
    notes = period_df.columns
    # bass =  bounds_df[notes[0]].values
    # last_note =  bounds_df[notes[-1]].values
    each_pair_values = []
    for i,row in bound_df.iterrows():
        bass_value = int(row['bass'])
        last_note_value = int(row['last_note'])
        mean_bound = row['mean']
        all_found, chosen_winners = binary_search_with_bounds(period_df,bass_value,last_note_value, mean_bound)
        

        ### interestingly enough, could i then here, normalize, and standard deviation?
        values=[row['bass_value']]
         ## indeces
        indices_to_observe=[row['bass']]
        for d in chosen_winners:
            values.append(d['candidate_value'])
            indices_to_observe.append(d['candidate_index'])
        indices_to_observe.append(row['last_note'])
        values.append(row['last_note_value'])


       


        ## normalizing, and calculating standard deviation and mean
        

        minimum = min(values)
        maximum = max(values)
        mean_vals = sum(values)/len(values)

        ##normalizing
        norm_vals = []
        # for v in values:
        #     normed = (v-minimum)/(maximum-minimum)
        #     norm_vals.append(normed)
        # std_= statistics.stdev(norm_vals)

        observable_metrics = {
            't_values':values,
            'mean_t_sub': mean_vals,
            # 'std_normed':std_,
            'indices':indices_to_observe,
            'difference':maximum-minimum,
            'difference_ratio': (maximum-minimum)/mean_vals,
        }

        




        each_pair_values.append({
            "all_found":all_found,
            'chosen_winners':chosen_winners,
            'observable_metrics':observable_metrics
            
        }
        )
        
    return each_pair_values
```

```python

# Function to generate sine wave data
def generate_sine_waves(frequencies, note_names, duration=1.0, sample_rate=1000):
    t = np.linspace(0, duration, int(sample_rate * duration))
    data = []
    
    for f_ind in range(len(frequencies)):
        freq = frequencies[f_ind]
        note_name = note_names[f_ind]
        # Calculate sine wave values: sin(2Ï€ * frequency * time)
        y_values = np.sin(2 * np.pi * freq * t)/(f_ind+1)  
        
        # Create DataFrame entries for this frequency
        for i in range(len(t)):
            data.append({
                'time': t[i],
                'amplitude': y_values[i],
                'frequency': f"{freq} Hz",
                'note_name':note_name
            })
            
    return pd.DataFrame(data)

# Generate the data
sine_df = generate_sine_waves(frequencies, note_names,duration=((1/frequencies[-1])*20), sample_rate=10000)
sine_df
```

```python
def generate_observable_df(note_names, fundamental_df, max_number_of_observable_metrics,max=20, metric_to_sort_by='difference_ratio'):
    period_df = calculate_period_dataframe(note_names,fundamental_df,max=max)
    bound_df = finding_min_difference_with_lower_upper_bound(period_df)
    bound_df.sort_values(by='ratio', ascending=True, inplace=True)
    each_pair = choosing_bounds_binary_search(period_df,bound_df)
    observable_metrics=[]
    for item in each_pair:
        observable_metrics.append(item['observable_metrics'])
    sorted_observable_metrics = sorted(observable_metrics, key=lambda x: x[metric_to_sort_by])
    return pd.DataFrame(sorted_observable_metrics[:max_number_of_observable_metrics])
```

```python
def create_chart_for_facet(observable_df, x='mean_t_sub',y='difference_ratio'):
    chart = alt.Chart(observable_df).mark_line().encode(
        x=alt.X(x, ),
        y=alt.Y(y, ),
        # color=alt.Color('difference_ratio', title='Difference Ratio'),
        tooltip=['t_values','mean_t_sub','std_normed','difference_ratio']
    ).properties(
        width=200,
        height=200
    )

    chart += alt.Chart(observable_df).mark_point().encode(
        x=alt.X(x, ),
        y=alt.Y(y, ),
        # color=alt.Color('difference_ratio', title='Difference Ratio'),
        tooltip=['t_values','mean_t_sub','std_normed','difference_ratio']
    ).properties(
        width=200,
        height=200
    )

    max_tsub = max(observable_df['mean_t_sub'])
    min_tsub = min(observable_df['mean_t_sub'])

    line_chart = alt.Chart(observable_df).mark_point().encode(
        y=alt.Y('difference_ratio'),
        color=alt.Color('mean_t_sub:Q', title='Mean Period', scale=alt.Scale(domain=[max_tsub, min_tsub])),
    )


    
    return (chart)
```

```python


def create_facet_chart(note_name_mega_array,titles , fundamental_df, max_number_of_observable_metrics,max=50,x='mean_t_sub',y='difference_ratio'):
    for i, note_names in enumerate(note_name_mega_array):
        observable_df = generate_observable_df(note_names, fundamental_df, max_number_of_observable_metrics,max=max)
        observable_df['title'] = titles[i]
        observable_df['row']= i//3
        if i == 0:
            all_observable_df = observable_df
        else:
            all_observable_df = pd.concat([all_observable_df, observable_df], ignore_index=True)

    all_observable_df['title'] = all_observable_df['title'].astype('category')
    return create_chart_for_facet(all_observable_df,x=x,y=y).facet(
        'title', align='each', columns=3), all_observable_df
 
```
