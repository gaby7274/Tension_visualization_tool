```python
import interharmonic_module.interharmonic_modulations as ihm
import altair as alt
import pandas as pd
import numpy as np

from scipy import stats

```

```python

def pipeline_chan_two_ranges_sum(number_of_harmonics=10,apply_weights_to_harmonics=True,limit_delta_frequency=True,include_lower_bound=False):
    # Example usage
    # Create harmonic dataframe

    df = ihm.create_harmonic_dataframe(harmonics_considered=number_of_harmonics)

    # Define notes to analyze
    notes_to_analyze = ['G3','G#3','A3', 'A#3', 'B3', 'C4', 'C#4', 'D4', 'D#4', 'E4', 'F4', 'F#4','G4']

    #define ranges for r_lower, r_upper
    values_of_r_ranges = [(0.95, 1.1),(1.5,2.8)]
    
    # Parameters for chan_tension_calculation
    no_weights = no_weights
    limit_delta_frequency = limit_delta_frequency
    include_lower_bound = include_lower_bound

    tension_list =[ihm.chan_tension_calculation('G3',note, df, values_of_r_ranges,apply_weights_to_harmonics=no_weights,limit_delta_frequency=limit_delta_frequency,include_lower_bound=include_lower_bound) for note in notes_to_analyze]
    # print(ihm.tension_df_print(tension_list, values_of_r_ranges))
    return tension_list, values_of_r_ranges
```

```python


def plot_interharmonic_plots_only_chan(number_of_harmonics_ranges,apply_weights_to_harmonics=True,limit_delta_frequency=True,include_lower_bound=False,empirical_ranking=[],interval_labels=['U', 'm2', 'M2', 'm3', 'M3', 'P4', 'TT', 'P5', 'm6', 'M6', 'm7', 'M7', '8v']):

    ## Biggest df of all results
    big_dataframe_for_ordering = pd.DataFrame()
    
    for number_of_harmonics in number_of_harmonics_ranges:
        tension_list, values_of_r_ranges = pipeline_chan_two_ranges_sum(number_of_harmonics,no_weights,limit_delta_frequency,include_lower_bound)
        # Create a DataFrame to hold the results
        # columns will be x:interval, y:tension_ranking, color:number_of_harmonics
        dataframe_for_ordering = pd.DataFrame(
            {'interval': interval_labels,
            'tension_ranking': [sum(tension) for tension in tension_list],
            'number_of_harmonics': [number_of_harmonics]*len(tension_list)
            }
        )
        big_dataframe_for_ordering = pd.concat([big_dataframe_for_ordering, dataframe_for_ordering], ignore_index=True)
    
    # Create the Altair plot
    chart = alt.Chart(big_dataframe_for_ordering).mark_line(point=True).encode(
        x=alt.X('interval:N', title='Interval', sort=empirical_ranking),
        y=alt.Y('tension_ranking:Q', title='Tension Ranking (Sum of Chan Tension)', scale=alt.Scale(reverse=True)),
        color=alt.Color('number_of_harmonics:N', title='Number of Harmonics', ),
        tooltip=['interval', 'tension_ranking', 'number_of_harmonics']
    ).properties(
        title='Interharmonic Modulation Tension Rankings by Number of Harmonics',
        width=600,
        height=400
    ).interactive()
    return chart


# define empirical ranking from reference 
empirical_ranking = ['U','8v','P5','P4','M3','M6','m6','m3','TT','m7','M2','M7','m2']


plot_interharmonic_plots_only_chan(number_of_harmonics_ranges=[4,8,16],apply_weights_to_harmonics=True,limit_delta_frequency=True,include_lower_bound=False,empirical_ranking=empirical_ranking)

```

 plan, for each 4,8,16,32,64 harmonics, get the tension ranking list and compare the list with the empirical ranking.
 make heatmap of correlations where each row is the number of harmonics and their correlation with empirical ranking

```python
import numpy as np
import scipy.stats as stats
empirical_ranking = ['U','8v','P5','P4','M3','M6','m6','m3','TT','m7','M2','M7','m2']


def compute_interharmonic_tension_correlation(number_of_harmonics_ranges=[4,8,16,32,64],apply_weights_to_harmonics=True,limit_delta_frequency=True,include_lower_bound=False,empirical_ranking=empirical_ranking,interval_labels=['U', 'm2', 'M2', 'm3', 'M3', 'P4', 'TT', 'P5', 'm6', 'M6', 'm7', 'M7', '8v']):
        ## Biggest df of all results
    # big_dataframe_for_ordering = pd.DataFrame()
    correlation_dataframe = pd.DataFrame()
    
    for number_of_harmonics in number_of_harmonics_ranges:
        tension_list, values_of_r_ranges = pipeline_chan_two_ranges_sum(number_of_harmonics,no_weights,limit_delta_frequency,include_lower_bound)
        # Create a DataFrame to hold the results
        # columns will be x:interval, y:tension_ranking, color:number_of_harmonics

        corr_df = ihm.return_correlation_df(tension_list, number_of_harmonics, empirical_ranking)
        correlation_dataframe = pd.concat([correlation_dataframe, pd.DataFrame(corr_df)], ignore_index=True)
    return correlation_dataframe
```

```python


def compute_interharmonic_tension_ranking_and_correlation(number_of_harmonics_ranges,apply_weights_to_harmonics=True,limit_delta_frequency=True,include_lower_bound=False,empirical_ranking=empirical_ranking):
    big_dataframe_for_ordering = pd.DataFrame()
    correlation_dataframe = pd.DataFrame()
    
    for number_of_harmonics in number_of_harmonics_ranges:
        tension_list, values_of_r_ranges = pipeline_chan_two_ranges_sum(number_of_harmonics,no_weights,limit_delta_frequency,include_lower_bound)
        # Create a DataFrame to hold the results
        # columns will be x:interval, y:tension_ranking, color:number_of_harmonics

        dataframe_for_ordering = ihm.return_tension_ranking_df(tension_list, number_of_harmonics, empirical_ranking=empirical_ranking)
        corr_df = ihm.return_correlation_df(tension_list, number_of_harmonics, empirical_ranking)
        
        big_dataframe_for_ordering = pd.concat([big_dataframe_for_ordering, dataframe_for_ordering], ignore_index=True)
        correlation_dataframe = pd.concat([correlation_dataframe, pd.DataFrame(corr_df)], ignore_index=True)
    return big_dataframe_for_ordering, correlation_dataframe
        
    
```

```python
def plot_normalized_interharmonic_plots_correlation_chan(big_dataframe_for_ordering,correlation_dataframe,empirical_ranking=[]):
    # Create the Altair plot for tension rankings
    chart = alt.Chart(big_dataframe_for_ordering).mark_line(point=True).encode(
        x=alt.X('interval:N', title='Interval', sort=empirical_ranking),
        y=alt.Y('tension_ranking:Q', title='Tension Ranking (Sum of Chan Tension)', scale=alt.Scale(reverse=True)),
        color=alt.Color('number_of_harmonics:N', title='Number of Harmonics', ),
        tooltip=['interval', 'tension_ranking', 'number_of_harmonics']
    ).properties(
        title='Unweighted Interharmonic Modulation Tension Rankings by Number of Harmonics (Normalized)',
        width=600,
        height=400
    ).interactive()
    
    # Create the Altair plot for correlation coefficients
    correlation_dataframe['correlation_coefficient'] = correlation_dataframe['correlation_coefficient'].round(3)
    corr_chart = alt.Chart(correlation_dataframe).mark_rect().encode(
        x=alt.X('type_of_modulation:N', title='Type of Modulation'),
        y=alt.Y('number_of_harmonics:O', title='Number of Harmonics'),
        color=alt.Color('correlation_coefficient:N', title='Correlation Coefficient', scale=alt.Scale(scheme='viridis')),
        tooltip=['number_of_harmonics', 'correlation_coefficient', 'type_of_modulation']
    ).properties(
        title='Correlation between Empirical and Computed Tension Rankings',
        width=600,
        height=400
    ).interactive()


    text_chart = alt.Chart(correlation_dataframe).mark_text(baseline='middle', color='black').encode(
        text='correlation_coefficient:N',
        y='number_of_harmonics:O',
        x='type_of_modulation:N'
    )
    corr_chart = corr_chart + text_chart
    
    return chart, corr_chart


```

```python
# define empirical ranking from reference 
empirical_ranking = ['U','8v','P5','P4','M3','M6','m6','m3','TT','m7','M2','M7','m2']

big_df, corr_df = compute_interharmonic_tension_ranking_and_correlation(number_of_harmonics_ranges=[4,8,12,16],apply_weights_to_harmonics=True,limit_delta_frequency=True,include_lower_bound=False,empirical_ranking=empirical_ranking)
ranking_linear_plot, corr_heatmap = plot_normalized_interharmonic_plots_correlation_chan(big_df,corr_df,empirical_ranking=empirical_ranking)
(ranking_linear_plot | corr_heatmap).resolve_scale(color='independent').resolve_legend(color='independent') 
```
