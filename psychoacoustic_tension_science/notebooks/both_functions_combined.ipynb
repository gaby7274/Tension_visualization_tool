```python
def multiply_by_weights(harmonic_methods, subharmonic_methods, weights={"ihm":0.61,'shm':0.39}):
    
    combined_df = pd.DataFrame()
    for i,r in harmonic_methods.iterrows():
        list_of_correlation = json.loads(r['corr_results_per_label'])

        
        for j, s in subharmonic_methods.iterrows():
            sub_list_of_correlation = json.loads(s['corr_results_per_label'])

            new_correlation = [(x *weights['ihm']) + (y*weights['shm']) for x,y in zip(list_of_correlation,sub_list_of_correlation)]
            new_corr_normalized = [(x-min(new_correlation))/(max(new_correlation)-min(new_correlation)) for x in new_correlation]

            temp_df = pd.DataFrame({
                'periods_considered': [s['periods_considered']],
                'harmonics_considered': [r['harmonics_considered']],
                'new_ranking': [json.dumps(new_correlation)],
                'new_rank_normalized': [json.dumps(new_corr_normalized)],
            })
            combined_df = pd.concat([combined_df, temp_df], ignore_index=True)
    return combined_df

combined_results_intervals = multiply_by_weights(ihm_chan_harm_intervals, shm_bp_intervals)
combined_results_intervals
```

```python
from scipy.stats import pearsonr
def get_correlation_between_empirical_and_method(empirical_ranking, df_with_rankings, numbered_ranking=None,chord_label_column='chord_label', column_to_compare_with_emp='raw_t_score_summed_only_modulations'):
    ## first normalize empirical ranking

    empirical_ranking_numbered = range(0, len(empirical_ranking))

    if numbered_ranking is None:
        normalized_values = [x / max(empirical_ranking_numbered) for x in empirical_ranking_numbered]
    else:
        mininum = min(numbered_ranking) 
        maximum = max(numbered_ranking)
        normalized_values = [(x - mininum) / (maximum - mininum) for x in numbered_ranking]

    ## we are going to create a dictionary with the chord labels
    # with their respective scores. The plan is to see how correlated they are. 
    keys_for_dict = df_with_rankings[chord_label_column].tolist()
    
    ranked_df = {}
    for key in keys_for_dict:
        ranked_df[key] = df_with_rankings[df_with_rankings[chord_label_column] == key][column_to_compare_with_emp].values[0]
    
    ## now we are going to construct a list based on the empirical ranking 
    # using this dictionary

    for chord in empirical_ranking:
        if chord not in ranked_df.keys():
            print(f'Warning: chord {chord} not found in the dataframe keys')
            continue
    values_to_correlate = [ranked_df[chord] for chord in empirical_ranking]

    # get correlation
    corr, p_value = pearsonr(normalized_values, values_to_correlate)

    return corr, p_value, values_to_correlate
```

```python
def break_into_periods_harmonics(df, empirical_ranking_labels=[],empirical_ranking_numbered =None,chord_column='chord_label',column_to_compare='normalized_value'):

    corr_df_for_p_h = pd.DataFrame()
    ## loop for each harmonic
    for harm in df.harmonics_considered.unique():
        print('Now processing ', harm,' harmonics')

        period_df = df[df['harmonics_considered']==harm]
        for period in period_df.periods_considered.unique():
            pdf = period_df[period_df['periods_considered']==period]


            correlation,p_value,scored_for_each_chord = shm.get_correlation_between_empirical_and_method(empirical_ranking=empirical_ranking_labels,
                                                                                                         df_with_rankings=pdf,
                                                                                                         numbered_ranking=empirical_ranking_numbered,
                                                                                                         chord_label_column=chord_column,
                                                                                                         column_to_compare_with_emp=column_to_compare)
            temp_df = pd.DataFrame({
                'correlation':[correlation],
                'p_value':[p_value],
                'scores_based_on_emp_rank':json.dumps(scored_for_each_chord),
                'periods_considered': [period],
                'harmonics_considered':[harm]

            })
            corr_df_for_p_h = pd.concat([corr_df_for_p_h, temp_df], ignore_index=True)

    return corr_df_for_p_h

corr_intervals_per_chord_label
```
